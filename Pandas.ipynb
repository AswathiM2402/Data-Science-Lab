#### 6. Scatter & Line Plots

Now you have to create the scatter plots and line plots between the following columns:

1. 'Finalmark` and `Attendance`

2. 'Finalmark` and `Series1`

3. 'Finalmark` and `Series2`

4. 'Finalmark` and `Attendance/10+Series1+Series2`




```
# Scatter plot between Finalmark and Attendance
import matplotlib.pyplot as plt
plt.scatter(df['FinalMark'], df['Attendance'])
plt.title('Finalmark and Attendance')
plt.xlabel('FinalMark')
plt.ylabel('Attendance')
plt.show()
```

```
# Line plot between Finalmark and Attendance

plt.plot(df['FinalMark'], df['Attendance'],linestyle='-')
plt.title('Finalmark and Attendance')
plt.xlabel('FinalMark')
plt.ylabel('Attendance')
plt.show()
```

```
# Scatter plot between Finalmark and Series1


plt.scatter(df['FinalMark'], df['Series1'], marker='o', linestyle='-')
plt.title('FinalMark and Series1')
plt.xlabel('FinalMark')
plt.ylabel('Series1')
plt.show()
```

```
# Line plot between Finalmark and Series1
plt.plot(df['FinalMark'], df['Series1'], marker='o', linestyle='-')
plt.title('FinalMark and Series1')
plt.xlabel('FinalMark')
plt.ylabel('Series1')
plt.show()
```

```
# Scatter plot between Finalmark and Series2


plt.scatter(df['FinalMark'], df['Series2'])
plt.title('Finalmark and Series2')
plt.xlabel('FinalMark')
plt.ylabel('Series2')
plt.show()
```

```
# Line plot between Finalmark and Series2

plt.plot(df['FinalMark'], df['Series2'],linestyle='-')
plt.title('Finalmark and Series2')
plt.xlabel('FinalMark')
plt.ylabel('Series2')
plt.show()
```

```
# Scatter plot between Finalmark and Series1+Series2+Attendance/10


plt.scatter(df['FinalMark'], df['Series2'],df['Series2'],df['Attendance'])
plt.title('Scatter plot between Finalmark and Series1+Series2+Attendance')
plt.xlabel('FinalMark')
plt.ylabel('Series1+Series2+Attendance')
plt.show()
```

```
# Line plot between Finalmark and Series1+Series2+Attendance/10

plt.plot(df['FinalMark'], df['Series2']+df['Series2']+df['Attendance']/10,linestyle='-')
plt.title('Scatter plot between Finalmark and Series1+Series2+Attendance')
plt.xlabel('FinalMark')
plt.ylabel('Series1+Series2+Attendance')
plt.show()
```

Plot Gerderwise count plot
